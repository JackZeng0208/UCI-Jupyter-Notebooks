# Finding Z-Values for Known Probabilities
## Question 1:
$x - N(30,4)$, Find $x_0$ such that $P(x \leq x_0) = 0.0062$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z_score = norm.ppf(0.0062)
print("Z-score =", z_score)
x = np.arange(10, 50, 0.001)
x_2 = np.arange(10, 20, 0.001)
plt.fill_between(x_2, norm.pdf(x_2, 30, 4), color='blue', alpha=0.5)
plt.plot(x, norm.pdf(x, 30, 4))
```

## Question 2:
$x - N(30,4)$ if $P(x \geq x_0) = 0.1056$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z_score = norm.ppf(0.1056) # find z-score for 0.1056(starting from -inf)
print("Z-score =", z_score)
x = np.arange(-3, 3, 0.001)
x_2 = np.arange(abs(z_score), 3, 0.001)
plt.fill_between(x_2, norm.pdf(x_2, 0, 1), color='blue', alpha=0.5)
plt.plot(x, norm.pdf(x, 0, 1))
```

## Question 3:
Find $x_0$ such that $P(x \geq x_0) = 0.8944$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

Actual_Prob = 1 - 0.8944
z_score = norm.ppf(Actual_Prob)
print("Z-score =", z_score)
x = np.arange(-3, 3, 0.001)
x_2 = np.arange(z_score, 3, 0.001)
plt.fill_between(x_2, norm.pdf(x_2, 0, 1), color='blue', alpha=0.5)
plt.plot(x, norm.pdf(x, 0, 1))
```

## Question 4:
Ambulance response time for station A is normally distributed with mean of 7.5 minutes and standard deviation of 2.5 minutes
1. Regulators require that 90% of all calls be reached in 9 minutes or less. Are the requirements met by this station?
- $P(X \leq 9) < 0.9$ -> then not met the requirements
- $z=\frac{x-\mu}{\sigma} = \frac{9-7.5}{2.5} = 0.6$
- $P(X \leq 0.6) = 0.7257$
- So, not met the requirement

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x_0 = 9
mean = 7.5
sd = 2.5
z_score = (x_0 - mean) / sd
print("Z-score =", z_score)
x = np.arange(0, 15, 0.001)
x_2 = np.arange(0, 9, 0.001)
print(norm(7.5, 2.5).pdf(9))
plt.fill_between(x_2, norm.pdf(x_2, 7.5, 2.5), color='blue', alpha=0.5)
plt.plot(x, norm.pdf(x, 7.5, 2.5))
```
